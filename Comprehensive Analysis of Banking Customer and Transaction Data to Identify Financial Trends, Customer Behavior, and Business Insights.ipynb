## Stage 2 – Data Cleaning and Pre-processing

- Handle missing values (impute or drop)

- Handle duplicates
- column name rename or dropping unwanted columns if required

- Check skewness and apply transformations

- Convert data types if needed

- Feature transformations (date parts, derived fields if required for analysis)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("https://raw.githubusercontent.com/arthi-rathinam/Main-Project/refs/heads/main/Banking%20Sector%20Anonymised%20Dataset%20-%20Raw%20-uncleaned.csv")
```

```python
# Handling duplicate values
df.duplicated().sum()
```

```python
# Handling missing values
df.isnull().sum()
```

```python
# Dropped the columns
print("\n Drop the Columns : ",df.drop(columns=['pincode', 'employer_name', 'relationship_manager_id','savings_interest_rate_pct', 'fd_interest_rate_pct',
                                                  'overdraft_limit_inr','num_products_held','branch_name','credit_bureau' ], inplace=True))
df.head(5)
```

```python
# Handling missing values
df['age'] = df['age'].fillna(df['age'].median())
```

```python
# Handling missing values
df['account_status'] = df['account_status'].fillna(df['account_status'].mode()[0])
df['gender'] = df['gender'].fillna(df['gender'].mode()[0])
```

```python
# Handling missing values - for Categorical col
cols = ['state', 'customer_segment', 'employment_status', 'kyc_status', 'nominee_added', 'mobile_banking_enabled',
        'net_banking_enabled', 'credit_card_linked', 'demat_account', 'risk_rating']

for col in cols:
  df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df['annual_income_inr'] = df['annual_income_inr'].fillna(df['annual_income_inr'].median())

df['account_balance_inr'] = df['account_balance_inr'].fillna(df['account_balance_inr'].median())

df['last_login_days_ago'] = df['last_login_days_ago'].fillna(df['last_login_days_ago'].median())
```

```python
binary_columns = ['is_dormant', 'churn_flag', 'fraud_flag','is_npa']

for col in binary_columns:
  df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df['credit_score'] = df['credit_score'].fillna(df['credit_score'].mean())

df['num_transactions_last_month'] = df['num_transactions_last_month'].fillna(df['num_transactions_last_month'].mean())
```

```python
df['loan_type'] = df['loan_type'].fillna('No Loan')

df['loan_amount_inr'] = df['loan_amount_inr'].fillna(0)

df['loan_tenure_months'] = df['loan_tenure_months'].fillna(0)

df['outstanding_principal_inr'] = df['outstanding_principal_inr'].fillna(0)

df['emi_amount_inr'] = df['emi_amount_inr'].fillna(0)
```

```python
df.loc[df['loan_type'] == 'No Loan', 'days_past_due'] = 0
df['days_past_due'] = df['days_past_due'].fillna(df['days_past_due'].mean())
```

```python
df.loc[df['loan_type'] == 'No Loan', 'penalty_amount_inr'] = 0
df['penalty_amount_inr'] = df['penalty_amount_inr'].fillna(df['penalty_amount_inr'].median())
```

```python
df['monthly_interest_income_inr'] = df['monthly_interest_income_inr'].fillna(df['monthly_interest_income_inr'].median())

df['monthly_service_charges_inr'] = df['monthly_service_charges_inr'].fillna(df['monthly_service_charges_inr'].median())

df['avg_transaction_value_inr'] = df['avg_transaction_value_inr'].fillna(df['avg_transaction_value_inr'].median())

df['last_transaction_amount_inr'] = df['last_transaction_amount_inr'].fillna(df['last_transaction_amount_inr'].median())
```

```python
df['gender'] = df['gender'].replace({'M' : 'Male', 'F' : 'Female'})
```

```python
df['account_open_date'] = pd.to_datetime(df['account_open_date'],format='%d-%m-%Y')
df['account_open_date'] = df['account_open_date'].dt.strftime('%d/%m/%Y')

df['last_transaction_date'] = pd.to_datetime(df['last_transaction_date'],format='%d-%m-%Y')
df['last_transaction_date'] = df['last_transaction_date'].dt.strftime('%d/%m/%Y')
```

```python
df.columns = df.columns.str.strip().str.capitalize()
```

```python
df.columns = df.columns.str.replace('inr', 'INR')
```

```python
df = df.drop_duplicates()
```
